```python
df.drop(columns=["customer_first_name", "customer_last_name", "days_for_shipment_scheduled", "days_for_shipment_real", "shipping_type", "customer_city", "customer_state", "ship_date"], inplace=True, axis=1)
```

```python
df['order_date']=pd.to_datetime(df['order_date'], errors='coerce')
```

```python
df
```

```python
df=df.dropna(subset=["order_date"])
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.copy()
df['Year'] = df['order_date'].dt.year
df['Month'] = df['order_date'].dt.month
df['Quarter'] = df['order_date'].dt.quarter
df['Month_Name'] = df['order_date'].dt.strftime('%B')
```

```python
df
```

```python
monthly_sales = df.groupby('Month_Name')['sales_per_order'].sum()
```

```python
month_order=["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
monthly_sales = monthly_sales.reindex(month_order)
```

```python
plt.figure(figsize=(14, 5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='purple', mec="black",ms="5")
plt.xlabel("Month",fontsize=12, fontweight='bold')
plt.ylabel("Total Sales",fontsize=12, fontweight='bold')
plt.title("Seasonal Sales Trends", fontsize=20, fontweight='bold')
plt.grid(True)
plt.show()
```
